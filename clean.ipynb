```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('weather_m4.csv')
df.shape
df.info()
df.head(5)
df.describe()
df.max()
df['PRESSURE'].min()
df['TEMP'].mode()
df['TEMP'][1] #row
#dft = df.T #change row to columns - transpose
df[['PRESSURE', 'TIME', 'TEMP']] #select columns, note brackets
df['TIME'] < 25 #condition
df[df['TIME'] >= df['TEMP']] # filtering record by condition
df['PASS'] = df.mean(axis=1) > df['TEMP'] # Creating a new column
df.sort_values(by='PASS') # sort by value
df.sort_values(by=['TIME', 'TEMP']) #sort by multiple value
df.agg(['min','max','mean', 'std'])
df.head(5)
```

```python
df.isnull().any() #Show all columns with null values
df[df.isnull().any(axis=1)] #show row with null values
#df.drop(columns='WEATHER_CODE', inplace=True) 
df.duplicated().any() #check for duplicate
df.drop_duplicates(inplace=True) #drop duplicates
df.dropna(inplace=True) #drop null values
rows_to_fill = df.isnull().any(axis=1) #rows with null values
df[rows_to_fill]
df['MIN_TEMP_GROUND'].fillna(method='bfill', inplace=True) #fill null values for column MIN_TEMP_GROUND
df[rows_to_fill]
rows_to_fill = df.isnull().any(axis=1) #more cleanup for null field
df[rows_to_fill]
nulls_filled = df.fillna(df.mean()) #fill null values with mean
nulls_filled[rows_to_fill]
```

```python
### remove outlier
# do not perform analysis on outlier dataset by removoving them
dft = df['TEMP']
#dft.plot.box()
q1 = dft.quantile(.25)
q3 = dft.quantile(.75)
iqr = q3 - q1 
pmin = q1 - 1.5 * iqr
pmax = q3 + 1.5 * iqr
nwh = dft.where(dft.between(pmin, pmax))
df.describe()
```

```python
#define function and apply
def iqr(col):
    q1 = col.quantile(.25)
    q3 = col.quantile(.75)
    return q3 - q1 


df.apply('TEMP')
```
